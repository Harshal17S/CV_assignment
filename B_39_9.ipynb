```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from ffpyplayer.player import MediaPlayer
```

```python
img=cv2.imread("astro.jpg")
img1=cv2.resize(img,(500,500))
operatedImage = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
operatedImage = np.float32(operatedImage)
dest =cv2.cornerHarris(operatedImage, 2, 5, 0.07)

threshold = 0.01 * dest.max()
marked_image = img1.copy()

for i in range(dest.shape[0]):
    for j in range(dest.shape[1]):
        if dest[i, j] > threshold:
            cv2.circle(marked_image, (j, i), 3, (0, 0, 255), -1)
plt.subplot(1, 3, 1)
plt.imshow(img1 )
plt.title('Original')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(operatedImage )
plt.title('Gray')
plt.axis('off')


plt.subplot(1, 3, 3)
plt.imshow(dest )
plt.title('Harris')
plt.axis('off')


plt.show()
```
