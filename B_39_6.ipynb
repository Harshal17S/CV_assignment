# Importing Libraries

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

# Reading the input image

```python
img = cv2.imread("astro.jpg", cv2.IMREAD_GRAYSCALE)
```

# Ploting histogram :

```python
Dog_hist = plt.hist(img.ravel(),256,[0,256])
cv2.waitKey(10000)
cv2.destroyAllWindows()
plt.show()
```

# Streaching an Image

```python
max_pix = np.amax(img)
min_pix = np.amin(img)

rows, cols = img.shape
img_linstr = ((img - min_pix) * (255.0 / (max_pix - min_pix))).astype('uint8')
cv2.imshow('OG Image', img)
cv2.imshow('Streached Image', img_linstr)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```

# Equalizing an Image

```python
eq_img = cv2.equalizeHist(img)
cv2.imshow('OG Image', img)
cv2.imshow('Equalized Image', eq_img)
cv2.waitKey(10000)
cv2.destroyAllWindows()
```
