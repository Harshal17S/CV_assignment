# Import Labarries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Translation

```python
img = cv2.imread("astro.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)                     
h, w = img.shape[:2]

x = int(input("Enter x-axis shift: "))
y = int(input("Enter y-axis shift: "))

t_matrix = np.float32([[1,0,x], [0,1,y]])                      
n_img = cv2.warpAffine(img, t_matrix, (w, h))

plt.imshow(img)
plt.show()
plt.imshow(n_img)
plt.show()
```

# Rotation

```python
img = cv2.imread("astro.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)                    
h, w = img.shape[:2]

center = (w//2, h//2)

ang = int(input("Enter degree of rotation: "))                  

r_matrix = cv2.getRotationMatrix2D(center, ang, 1.0)            
r = cv2.warpAffine(img, r_matrix, (w, h))

plt.imshow(img)
plt.show()
plt.imshow(r)
plt.show()
```

# Scaling

```python
img = cv2.imread("astro.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)                     
h, w = img.shape[:2]

scale = float(input("Enter scaling factor: "))                  

h = int(img.shape[0] * scale)                                   
w = int(img.shape[1] * scale)                                   
re = cv2.resize(img, (w, h))                                    

plt.imshow(img)
plt.show()
plt.imshow(re)
plt.show()
```

# Cropping

```python
img = cv2.imread("astro.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)                     
h, w = img.shape[:2]

print(img.shape)

cxlow = int(input("Enter lower-x limit: "))                    
cxup = int(input("Enter upper-x limit: "))
cylow = int(input("Enter lower-y limit: "))
cyup = int(input("Enter upper-y limit: "))

crop = img[cxlow:cxup, cylow:cyup]                             

plt.imshow(img)
plt.show()
plt.imshow(crop)
plt.show()
```

# Flipping using CV2.flip

```python
img = cv2.imread("astro.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)                     
h, w = img.shape[:2]

flip1 = cv2.flip(img, 1)
flip2 = cv2.flip(img, 0)
flip3 = cv2.flip(img, -1)

plt.imshow(img)
plt.show()
plt.imshow(flip1)
plt.show()

plt.imshow(flip2)
plt.show()

plt.imshow(flip3)
plt.show()

```

# Flipping using warperspective transform

```python

img = cv2.imread("astro.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)                     
h, w = img.shape[:2]

m = int(input("Enter 1 for horizontal, 2 for vertical or 3 for both axes flipping: "))
if m==1:
    matrix = np.float32([[-1,0,w], [0,1,0], [0,0,1]])         
    flip = cv2.warpPerspective(img, matrix, (w,h))
elif m==2:
    matrix = np.float32([[1,0,0], [0,-1,h], [0,0,1]])         
    flip = cv2.warpPerspective(img, matrix, (w,h))
elif m==3:
    matrix = np.float32([[-1,0,w], [0,-1,h], [0,0,1]])        
    
    flip = cv2.warpPerspective(img, matrix, (w,h))

plt.imshow(img)
plt.show()
plt.imshow(flip)
plt.show()
```

# Shearing

```python

img = cv2.imread("astro.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)                     
h, w = img.shape[:2]

m = int(input("Enter 1 for horizontal and 2 for vertical shear: "))

if m==1:
    k = float(input("Enter shearing factor: "))
    sh_matrix = np.float32([[1,k,0], [0,1,0], [0,0,1]])                  
    sheared = cv2.warpPerspective(img, sh_matrix, (int(w*(k+1)), h))    

if m==2:
    k = float(input("Enter shearing factor: "))
    sh_matrix = np.float32([[1,0,0], [k,1,0], [0,0,1]])
    sheared = cv2.warpPerspective(img, sh_matrix, (w, int(h*(k+1))))     

plt.imshow(img)
plt.show()
plt.imshow(sheared)
plt.show()
```
