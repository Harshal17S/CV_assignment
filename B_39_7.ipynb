# Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from ffpyplayer.player import MediaPlayer
import time
```

# Low Pass Average Filter

```python
img = cv2.imread("astro.jpg", 0)
img1=cv2.resize(img,(400,400))
m, n= img1.shape
mask = np.ones([3, 3], dtype = int)
mask =mask/9
img_new=cv2.filter2D(img1, -1,mask)
img_new= img_new.astype(np.uint8)
plt.subplot(1, 2, 1)
plt.imshow(img1, cmap='gray')
plt.title('Original')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(img_new, cmap='gray')
plt.title('Low Pass Average Filter')
plt.axis('off')
plt.show()
```

# Low Passs Median Filter

```python
img_noisy1 = cv2.imread('astro.jpg', 0)
img1 = cv2.resize(img_noisy1, (500, 400))
img_new1 = cv2.medianBlur(img1, 5)
plt.subplot(1, 2, 1)
plt.imshow(img1, cmap='gray')
plt.title('Original')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(img_new1, cmap='gray')
plt.title('Low Pass Median Filter')
plt.axis('off')
plt.show()
```

# High Pass Filter

```python
img=cv2.imread("astro.jpg", 0)
img1=cv2.resize(img_noisy1,(500,500))
m, n=img1.shape
kernel_high_pass = np.array([
                            [-1, -1, -1],
                            [-1,  8, -1], 
                            [-1, -1, -1]
                            ], np.float32) 
img_high_pass = cv2.filter2D(img1, -1, kernel_high_pass)
plt.subplot(1, 2, 1)
plt.imshow(img1, cmap='gray')
plt.title('Original')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(img_high_pass, cmap='gray')
plt.title('High Pass Filter')
plt.axis('off')
plt.show()
```

# High Boost Filter

```python
img = cv2.imread('astro.jpg', 0)
k = 2
img_smooth = cv2.GaussianBlur(img, (5, 5), 0)
img_highpass = img - img_smooth
img_highboost = img + k * img_highpass
plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')
plt.subplot(1, 2, 2)
plt.imshow(img_highboost, cmap='gray')
plt.title('High Boost Filter')
plt.axis('off')
plt.show()
```
