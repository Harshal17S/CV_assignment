# Import Libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
from ffpyplayer.player import MediaPlayer
import time
```

# creating Roberts Operator

```python
img = cv2.imread("lena.jpeg", 0)
img1 = cv2.resize(img, (500, 400))
kernelx = np.array([[0, 1], [-1, 0]])
kernely = np.array([[1, 0], [0, -1]])
img_robertx = cv2.filter2D(img1, -1, kernelx)
img_roberty = cv2.filter2D(img1, -1, kernely)
plt.subplot(1, 3, 1)
plt.imshow(img1, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(img_robertx, cmap='gray')
plt.title('Robertx')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(img_roberty, cmap='gray')
plt.title('Roberty')
plt.axis('off')

plt.show()
```

# Prewitt's Operator

```python
img = cv2.imread("lena.jpeg", 0)
img1 = cv2.resize(img, (1000, 700))
kernelx = np.array([[-1, 0, 1], [-1, 0, 1], [-1, 0, 1]])
kernely = np.array([[1, 1, 1], [0, 0, 0], [-1, -1, -1]])
img_prewittx = cv2.filter2D(img1, -1, kernelx)
img_prewitty = cv2.filter2D(img1, -1, kernely)
plt.subplot(1, 3, 1)
plt.imshow(img1, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(img_prewittx, cmap='gray')
plt.title('Prewittx')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(img_prewitty, cmap='gray')
plt.title('Prewitty')
plt.axis('off')
plt.show()
```

# Sobel Operator

```python
img = cv2.imread("lena.jpeg", 0)
img1 = cv2.resize(img, (1000, 700))
kernelx = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
kernely = np.array([[1, 2, 1], [0, 0, 0], [-1, -2, -1]])
img_sobelx = cv2.filter2D(img1, -1, kernelx)
img_sobely = cv2.filter2D(img1, -1, kernely)
plt.subplot(1, 3, 1)
plt.imshow(img1, cmap='gray')
plt.title('Original')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(img_sobelx, cmap='gray')
plt.title('Sobelx')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(img_sobely, cmap='gray')
plt.title('Sobely')
plt.axis('off')

plt.show()
```
