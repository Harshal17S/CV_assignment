```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# AND operation


```python
img1 = cv2.imread("astro.jpg")
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2GRAY)
img2 = cv2.imread("nature.jpg")
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
print(img1.shape)
print(img2.shape)

# resixe 10 x10
width = 10
height = 10 
dim = (width, height)
  
# resize image
imgR1 = cv2.resize(img1, dim)
imgR2 = cv2.resize(img2, dim)

AND = cv2.bitwise_and(imgR1, imgR2)

plt.imshow(imgR1, cmap='gray')
plt.show()
print(imgR1)
plt.imshow(imgR2, cmap='gray')
plt.show()
print(imgR2)
plt.imshow(AND, cmap='gray')
plt.show()
print(AND)
```

# NOT Operation
